```python
import matplotlib.pyplot as plt
import numpy as np
```

Resolver por transformada inversa:
$$f(x)=\left\{\begin{matrix}
x & si \quad   0\leq x\leq 1  \\
 \frac{1}{2}& si\quad  1\leq x\leq 2
\end{matrix}\right.$$

La funcion $f(x)$ es una f.d.p, pues:
$$f(x)\geq 0$$
y
$$\int_{0}^{2}f(x)dx=\int_{0}^{1}xdx+\int_{1}^{2}\frac{1}{2}dx= \frac{x^2}{2}\bigg|_{0}^{1}+\frac{x}{2}\bigg|_{1}^{2}=1$$
Definimos $F(x)$ como:
$$F(x)=\left\{\begin{matrix}
\frac{x^2}{2} &, 0\leq x\leq 1 \\
\frac{x}{2} &, 1\leq x\leq 2
\end{matrix}\right.$$

Transformada inversa, resolvemos: $$F(x)=R$$
$$\frac{x^2}{2}=R$$
$$x= \sqrt{2R} ,\quad si \quad 0\leq u \leq \frac{1}{2}$$
Y
$$x=2R,\quad si \quad u\geq\frac{1}{2}$$


### Grafica f.d.p:

```python
def f(x):
  if 0<=x<=1:
    return x
  elif 1<=x<=2:
    return 1/2
```

```python
X=np.linspace(0, 2, 400) #Generamos numeros aleatorios
Y=[f(x) for x in X]      #Evaluamos los numeros aleatorios generados

plt.plot(X,Y)
plt.grid(True)  # Mallamos la grafica
plt.ylim(0, 1)  # ajustar el límite en y
plt.xlim(0, 2)  # ajustar el límite en x
plt.show()
```

###Grafica acumulativa:


```python
def F(x):
  if 0<=x<=1:
    return x**2/2
  elif 1<=x<=2:
    return x/2
```

```python
X=np.linspace(0, 2, 400)
Y=[F(x) for x in X]

plt.plot(X,Y)
plt.grid(True)
plt.ylim(0, 1)
plt.xlim(0, 2)
plt.show()
```

###Histograma de la muestra:

```python
def F1(x):
  if 0 <= x <= 0.5:
    return (2*x)**(1/2)
  elif x >= 1/2:
    return 2*x
```

```python
X=np.linspace(0, 2, 400)
Y=[F1(x) for x in X]

plt.hist(Y,bins=100)
plt.grid(True)
plt.show()
```
